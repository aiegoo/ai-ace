### ì‹¤ìŠµ 5

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv("mpg.csv")

# categoryë³„ë¡œ cty, hwy í‰ê·  ê³„ì‚° í›„ chart ë³€ìˆ˜ì— ì €ì¥
chart = df.groupby("category")[["cty", "hwy"]].mean()

# ì„  ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
plt.figure(figsize=(8, 5))
plt.plot(chart.index, chart["cty"], marker='o', label='ë„ì‹œ ì—°ë¹„ (cty)', color='blue')
plt.plot(chart.index, chart["hwy"], marker='s', label='ê³ ì†ë„ë¡œ ì—°ë¹„ (hwy)', color='green')

# ê·¸ë˜í”„ ê¾¸ë¯¸ê¸°
plt.title("ì°¨ì¢…ë³„ í‰ê·  ì—°ë¹„")
plt.xlabel("ì°¨ì¢… (category)")
plt.ylabel("ì—°ë¹„ (mpg)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

# ê·¸ë˜í”„ ì¶œë ¥
plt.show()
```
