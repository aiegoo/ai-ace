```python
# í•„ìˆ˜ ë¼ì´ë¸ŒëŸ¬ë¦¬ Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Scikit-Learn ëª¨ë“ˆë“¤
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit, TimeSeriesSplit
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

# í•œê¸€ í°íŠ¸ ì„¤ì •
def setup_korean_fonts():
    """í•œê¸€ í°íŠ¸ë¥¼ ìë™ìœ¼ë¡œ ì„¤ì •í•˜ëŠ” í•¨ìˆ˜"""
    import platform
    import matplotlib.font_manager as fm
    
    print("ğŸ”¤ í•œê¸€ í°íŠ¸ ì„¤ì • ì¤‘...")
    
    # ì‹œìŠ¤í…œë³„ í•œê¸€ í°íŠ¸ ë¦¬ìŠ¤íŠ¸
    if platform.system() == 'Windows':
        font_candidates = ['Malgun Gothic', 'Microsoft YaHei', 'SimHei', 'Arial Unicode MS']
    elif platform.system() == 'Darwin':  # macOS
        font_candidates = ['AppleGothic', 'Nanum Gothic', 'Arial Unicode MS']
    else:  # Linux
        font_candidates = ['Nanum Gothic', 'DejaVu Sans', 'Liberation Sans']
    
    # ì‚¬ìš© ê°€ëŠ¥í•œ í°íŠ¸ ì°¾ê¸°
    available_fonts = [f.name for f in fm.fontManager.ttflist]
    print(f"ğŸ“ ì‹œìŠ¤í…œ í°íŠ¸ ìˆ˜: {len(available_fonts)}ê°œ")
    
    selected_font = None
    for font in font_candidates:
        if font in available_fonts:
            selected_font = font
            break
    
    if selected_font:
        plt.rcParams['font.family'] = selected_font
        plt.rcParams['axes.unicode_minus'] = False  # ë§ˆì´ë„ˆìŠ¤ ë¶€í˜¸ ê¹¨ì§ ë°©ì§€
        print(f"âœ… ì„ íƒëœ í•œê¸€ í°íŠ¸: {selected_font}")
        return selected_font
    else:
        print("âš ï¸ í•œê¸€ í°íŠ¸ë¥¼ ì°¾ì„ ìˆ˜ ì—†ì–´ ê¸°ë³¸ í°íŠ¸ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.")
        return "DejaVu Sans"

# í•œê¸€ í°íŠ¸ ì„¤ì • ì‹¤í–‰
font_name = setup_korean_fonts()
print(f"âœ… í•œê¸€ í°íŠ¸ ì„¤ì • ì™„ë£Œ: {font_name}")

# ê¸°ë³¸ ì„¤ì •
plt.style.use('default')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', 50)

print("âœ… ëª¨ë“  ë¼ì´ë¸ŒëŸ¬ë¦¬ê°€ ì„±ê³µì ìœ¼ë¡œ ë¡œë“œë˜ì—ˆìŠµë‹ˆë‹¤!")
print(f"ğŸ“Š Pandas Version: {pd.__version__}")
print(f"ğŸ”¢ NumPy Version: {np.__version__}")
print(f"ğŸ¤– Scikit-Learn ë¶„í•  ëª¨ë“ˆë“¤ì´ ì¤€ë¹„ë˜ì—ˆìŠµë‹ˆë‹¤!")
print(f"ğŸ“ˆ Matplotlib & Seaborn ì‹œê°í™” ì¤€ë¹„ì™„ë£Œ!")
print(f"ğŸ”¤ í•œê¸€ í°íŠ¸: {font_name}")

print("\nğŸš— MPG ìë™ì°¨ ì—°ë¹„ ë°ì´í„° ë¶„ì„ì„ ì‹œì‘í•˜ê² ìŠµë‹ˆë‹¤!")
```

### ì‹¤ìŠµ 5

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_csv("../../data/mpg.csv")

# categoryë³„ë¡œ cty, hwy í‰ê·  ê³„ì‚° í›„ chart ë³€ìˆ˜ì— ì €ì¥
chart = df.groupby("category")[["cty", "hwy"]].mean()

# ì„  ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
plt.figure(figsize=(8, 5))
plt.plot(chart.index, chart["cty"], marker='o', label='ë„ì‹œ ì—°ë¹„ (cty)', color='blue')
plt.plot(chart.index, chart["hwy"], marker='s', label='ê³ ì†ë„ë¡œ ì—°ë¹„ (hwy)', color='green')

# ê·¸ë˜í”„ ê¾¸ë¯¸ê¸°
plt.title("ì°¨ì¢…ë³„ í‰ê·  ì—°ë¹„")
plt.xlabel("ì°¨ì¢… (category)")
plt.ylabel("ì—°ë¹„ (mpg)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

# ê·¸ë˜í”„ ì¶œë ¥
plt.show()
```
